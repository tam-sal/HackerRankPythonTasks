```python
n = int(input('user input: ').strip())
Constraints = list(range(1, 101))
case_1 = n%2==1
case_2 = list(filter(lambda item: item % 2 == 0, list(range(2,6))))
case_3 = list(filter(lambda item: item % 2 == 0, list(range(6,21))))
case_4 = n%2 == 0 and n > 20
if n in Constraints:

    if case_1:
        print("Weird")
    elif n in case_2:
        print('Not Weird')
    elif n in case_3:
        print("Weird")
    elif case_4:
        print('Not Weird')
else:
    print("Not valid")
```

```python
def is_leap(year):
    try:
        year in range(1900, 10**5+1)
    except Exception as e:
        e = 'year out of range'
        return e
        
    leap = False
    
    conLeap_1 = year % 4 == 0
    conLeap_2 = conLeap_1 and not(year % 100 == 0)
    conLeap_3 = conLeap_1 and not(conLeap_2) and year % 400 == 0
    if conLeap_2 or conLeap_3:
        leap = True
    else:
        leap = False
    
    return leap
```

```python
import re
def valid_card(card_number):
    pattern1 = re.compile('\d{16}')
    pattern2 = re.compile('\d{4}-\d{4}-\d{4}-\d{4}')
    repeat_pat = re.compile("\\b([0-9])\\1\\1+\\1+\\b")
    valid_1 = bool(re.match(pattern1, card_number))
    valid_2 = bool(re.match(pattern2, card_number))
    valid_start = ['4','5','6']
    if valid_1 or valid_2:
        pass
    else:
        return('Invalid')
    
    card_number_list = [item for item in card_number if item != '-']
    if card_number_list[0] not in valid_start or len(card_number_list) != 16:
        return('Invalid')
    for i in range(len(card_number_list)):
        sample_4 = ''.join(card_number_list[i:i+4])
        is_repeated = bool(re.match(repeat_pat, sample_4))
        if is_repeated:
            return('Invalid')
    return ('Valid')


nbr_of_cards = int(input(''))
for card in range(nbr_of_cards):
    card_number = input('').strip()
    print(valid_card(card_number))
```
